```python
import tkinter as tk
from tkinter import messagebox
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

# Function to perform auto-login
def auto_login():
    url = url_entry.get()
    username = username_entry.get()
    password = password_entry.get()
    login_button_xpath = login_button_entry.get()

    if not url or not username or not password or not login_button_xpath:
        messagebox.showwarning("Input Error", "All fields are required!")
        return

    try:
        # Set up the Chrome driver
        driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
        driver.get(url)

        # Find the username and password fields and enter the credentials
        driver.find_element(By.NAME, 'username').send_keys(username)
        driver.find_element(By.NAME, 'password').send_keys(password)

        # Find the login button and click it
        driver.find_element(By.XPATH, login_button_xpath).click()
        messagebox.showinfo("Success", "Login Successful!")
    except Exception as e:
        messagebox.showerror("Error", f"An error occurred: {e}")
    finally:
        driver.quit()

# Setting up the Tkinter window
root = tk.Tk()
root.title("Auto-Login Bot")
root.geometry("400x300")
root.resizable(False, False)
root.configure(bg="#2c3e50")

# Styling
label_font = ("Arial", 12, "bold")
entry_font = ("Arial", 12)
button_font = ("Arial", 12, "bold")

# URL
url_label = tk.Label(root, text="URL", font=label_font, bg="#2c3e50", fg="#ecf0f1")
url_label.pack(pady=(20, 5))
url_entry = tk.Entry(root, font=entry_font, width=30)
url_entry.pack(pady=5)

# Username
username_label = tk.Label(root, text="Username", font=label_font, bg="#2c3e50", fg="#ecf0f1")
username_label.pack(pady=5)
username_entry = tk.Entry(root, font=entry_font, width=30)
username_entry.pack(pady=5)

# Password
password_label = tk.Label(root, text="Password", font=label_font, bg="#2c3e50", fg="#ecf0f1")
password_label.pack(pady=5)
password_entry = tk.Entry(root, font=entry_font, width=30, show="*")
password_entry.pack(pady=5)

# Login Button XPath
login_button_label = tk.Label(root, text="Login Button XPath", font=label_font, bg="#2c3e50", fg="#ecf0f1")
login_button_label.pack(pady=5)
login_button_entry = tk.Entry(root, font=entry_font, width=30)
login_button_entry.pack(pady=5)

# Auto-login Button
login_button = tk.Button(root, text="Login", font=button_font, bg="#e74c3c", fg="#ecf0f1", command=auto_login)
login_button.pack(pady=20)

# Start the GUI event loop
root.mainloop()
```
